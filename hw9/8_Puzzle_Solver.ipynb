```python
import heapq
import copy

def print_succ(state):

    # Create an empty list of successor states:
    successors = []
    heuristics = []

    # Get the index of the blank space in the grid:
    zero = state.index(0)

    # UP: if zero not on top row, swap it with tile above it
    if zero > 2:
        newState = copy.deepcopy(state)
        newState[zero] = newState[zero - 3]
        newState[zero - 3] = 0
        successors.append(newState)

    # DOWN: If zero not on bottom row, swap it with tile below it
    if zero < 6:
        newState = copy.deepcopy(state)
        newState[zero] = newState[zero + 3]
        newState[zero + 3] = 0
        successors.append(newState)

    # LEFT: If zero not in left column, swap it with tile to the left
    if zero % 3 > 0:
        newState = copy.deepcopy(state)
        newState[zero] = newState[zero - 1]
        newState[zero - 1] = 0
        successors.append(newState)

    # RIGHT: If zero not on right column, swap it with tile to the right
    if zero % 3 < 2:
        newState = copy.deepcopy(state)
        newState[zero] = newState[zero + 1]
        newState[zero + 1] = 0
        successors.append(newState)

    # Sort the list of successors, then return:
    successors.sort()

    for i in successors:
        h = manhattan_dist(i)
        heuristics.append(h)

    return successors

def index(item, seq):
    
    if item in seq:
        return seq.index(item)
    
    else:
        return -1

def manhattan_dist(state):

    h = sum(abs((val - 1) % 3 - i %3) + abs((val - 1) // 3 - i // 3)
        for i, val in enumerate(state) if val)

    return h 

def is_solved(state):
        goal_state = [1,2,3,4,5,6,7,8,0]
        current_state = state
        if current_state == goal_state:
            return True

        else:
            return False
        
########################################################################################################

def solve(state):
    """
    Function to solve the 8 tile puzzle. Uses the helper functions 'is_solved' and 
    'manhattan_dist' in order to get heuristics on the current state of the puzzle.
    If the puzzle is not currently in the goal state, the A* search will continue and attempt 
    to solve the puzzle.
    """

    # Setup the open and close priotity queues, set move counter to zero:
    open_queue = []
    closed_queue = []
    move_count = 0
    parent_index = -1

    # Place the starting node in the priority queue:
    heapq.heappush(open_queue, (move_count + manhattan_dist(state), state, (move_count, manhattan_dist(state), parent_index)))

    # If the queue is empty, exit. Otherwise, we shuffle the node into the closed queue:
    while (is_solved(state) == False):

        state = heapq.heappop(open_queue)
        heapq.heappush(closed_queue, state)
        parent_index += 1

        # Is the puzzle solved?
        if (is_solved(state) == True):
            # Yes
            print('Solved')

        # No:
        else:

            # Get the list of successor states:
            successors = print_succ(state[1])

            move_count += 1

            # Have we seen these moves before:
            for move in successors:
                idx_open = idx_closed = -1

                idx_open = index(move, open_queue)
                idx_closed = index(move, closed_queue)

                if ((idx_open and idx_closed) == -1):
                    heapq.heappush(open_queue, (move_count + manhattan_dist(move), move, (move_count, manhattan_dist(move), parent_index)))

solve([4,3,8,5,1,6,7,2,0])
```
